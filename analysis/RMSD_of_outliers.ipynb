```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_fns = list(map(lambda x: f"outliers_{x}.csv", range(25)))
```

```python
frames = []
```

```python
for csv_fn in csv_fns:
    df = pd.read_csv(csv_fn, names=['md5','RMSD'], header=0)
    frames.append(df)
```

```python
fig, axes = plt.subplots(nrows=len(frames), ncols=1, figsize=(4, 45))
fig.tight_layout(pad=3)
i = 0
for fr in frames:
    fr.RMSD.hist(ax=axes[i],log=True); axes[i].set_xlim([5,10.5]); axes[i].set_xlabel('RMSD'); 
    axes[i].set_title(f"Outlier search iteration {i}"); 
    i += 1
```

```python
# number of outliers in different iterations

print(list(map(lambda x: len(x), frames)))
```
